```python
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
# 1. Загрузить файл HR.csv в pandas dataframe
print('Задание 1\n')
data = pd.read_csv('HR.csv')
display(data)
```

```python
# 2. Рассчитайте основные статистики для переменных
#    (среднее, медиана, мода, мин/макс, стандартное отклонение).
print('Задание 2\n')
results_df = pd.DataFrame({'Stat. names' : ['mean','mediana', 'mode', 'min' , 'max', 'std']})

my_list = data.columns.tolist()
my_list.remove('department')
my_list.remove('salary')

for n in my_list:
    results_df[n] = [data[n].mean(axis = 0) , data[n].median(), data[n].mode()[0], data[n].min(), data[n].max(), data[n].std()]

display(results_df)  
```

```python
# 3. Рассчитайте и визуализируйте корреляционную матрицу для
#    количественных переменных.
#    Определите две самые скоррелированные и две наименее
#    скоррелированные переменные.
print('Задание 3\n')
corr = data.corr()
display(corr)
corr.style.background_gradient(cmap='coolwarm')
plt.matshow(corr)

plt.show()
print('Наиболее скоррелированные переменные: number_project и average_montly_hours')
print('Наименее скоррелированные переменные: left и satisfaction_level')
```

```python
# 4. Рассчитайти сколько сотрудников работает в каждом департаменте
print ('Задание 4\n')
ddf = data['department'].value_counts()
print(ddf)

```

```python
# 5. Показать распределение сотрудников по зарплатам.
print ('\nЗадание 5')

vals = data['salary'].values
uniq, idx = np.unique(vals, return_inverse=True)
ddff = data.groupby(['salary'])['department'].count() 

dataf = ddff.to_frame()
dataf = dataf.rename(columns={'department' : 'number_of_employees'})
dataf['salary_type'] = [0, 1, 2]

plt.bar(dataf['salary_type'], dataf['number_of_employees'], width=0.3)

plt.xticks([0,1,2], uniq)
plt.xlabel('Зарплата')
plt.ylabel('Количество сотрудников')
plt.show()


```

```python
# 6. Показать распределение сотрудников по зарплатам в каждом департаменте по отдельности
print ('Задание 6\n')
departments = data['department'].unique()
#print(departments) # получили список отделов
for dep in departments:
    print('Для отдела',dep)
    ddff = data[data['department'] == dep]
    ddff = ddff.groupby(['salary'])['department'].count()
    dataf = ddff.to_frame()
    dataf = dataf.rename(columns={'department' : 'number_of_employees'})
    dataf['salary_type'] = [0, 1, 2]
    dataf = dataf.rename(columns={'department' : 'number_of_employees'})
    dataf['salary_type'] = [0, 1, 2]
    plt.bar(dataf['salary_type'], dataf['number_of_employees'], width=0.3)
    plt.xticks([0,1,2], uniq)
    plt.xlabel('Зарплата')
    plt.ylabel('Количество сотрудников')
    plt.show()
    
```

```python
# 7. Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом
print ('Задание 7\n')

ddd = data[(data['salary'] == 'high') | (data['salary'] == 'low')]
ddd = ddd.reset_index()
ddd = ddd.drop('index', axis = 1)
display(ddd)

x = ddd[ddd['salary'] == 'high']['average_montly_hours']
y = ddd[ddd['salary'] == 'low']['average_montly_hours']
x.name, y.name = 'high', 'low'

def two_hist(x,y):
    x.hist(alpha = 0.5, weights = [1./len(x)]*len(x))
    y.hist(alpha = 0.5, weights = [1./len(y)]*len(y))
    plt.axvline(x.mean(), color='red', alpha=0.8, linestyle='dashed')
    plt.axvline(y.mean(), color='blue', alpha=0.8, linestyle='dashed')
    plt.legend([x.name, y.name])
    plt.show()
    
two_hist(x,y)
print('Оба распределения Гауссовы.')
print('Подтвердим это используя тест Шапиро-Вилкса:')

from scipy.stats import shapiro

stat, p = shapiro(x)
print('p-val для high =' , p)
print('Т.к. p-val < 0.05, то распределение нормальное')
stat, p = shapiro(y)
print('p-val для low =' , p)
print('Т.к. p-val < 0.05, то распределение нормальное')
#распределение нормальное,значит можно использовать любой критерий

from scipy.stats import mannwhitneyu
#даже если бы распределение не было бы нормальным
stat, p = mannwhitneyu(x,y)
print('Применим критерий Манна-Уитни')
print('p-val =' , p)
print('Т.к. p-val значительно больше уровня значимости = 0.05, то есть сильная зависимость')
print('Значит нет разницы во времени работы в месяц между сотрудниками с низкой и высокой зарплатой')
```

```python
# 8. Рассчитать следующие показатели среди уволившихся и неуволившихся сотрудников (по отдельности):
# - Доля сотрудников с повышением за последние 5 лет
# - Средняя степень удовлетворенности
# - Среднее количество проектов
print ('Задание 8 \n')
employee_left = data[data['left'] == 1]
employee_work = data[data['left'] == 0]

print('Среди уволившихся:\n')
print('Доля сотрудников с повышением за последние 5 лет =', employee_left['promotion_last_5years'].value_counts()[1]/employee_left.shape[0])
print('Средняя степень удовлетворенности =', employee_left['satisfaction_level'].mean())
print('Среднее количество проектов =', employee_left['number_project'].mean())

print('\nСреди работающих:\n')
print('Доля сотрудников с повышением за последние 5 лет =', employee_work['promotion_last_5years'].value_counts()[1]/employee_work.shape[0])
print('Средняя степень удовлетворенности =', employee_work['satisfaction_level'].mean())
print('Среднее количество проектов =', employee_work['number_project'].mean())

#display(employee_left)
```

```python
# 9. Разделить данные на тестовую и обучающую выборки. Построить модель LDA, предсказывающую уволился 
# ли сотрудник на основе имеющихся факторов (кроме department и salary)
# Оценить качество модели на тестовой выборки

from sklearn.model_selection import train_test_split

print ('Задание 9.2\n')

data1 = data.drop('department', axis=1)
data1 = data1.drop('salary', axis=1)
display(data1)
y = data1['left']
X_train, X_test, y_train, y_test = train_test_split(data1, y, test_size=0.30, random_state=42)

#display(X_train)
#display(X_test)

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

model = LinearDiscriminantAnalysis()
# обучаем модель
model.fit(X_train, y_train)

# предсказываем
yhat = model.predict(X_test)

display(X_test)
display(yhat)


```
