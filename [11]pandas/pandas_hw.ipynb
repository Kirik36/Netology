## Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd

#для датасета ml-latest-small.zip

data = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

```python
movie_name = movies[movies['movieId'] == data[data['rating']==5]['movieId'].value_counts().head(1).iloc[0]]['title'].iloc[0]
print('Больше всего оценок 5.0 получил фильм:',movie_name)
```

## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head(10)
```

```python
countries = ['Latvia', 'Lithuania', 'Estonia']

for current_country in countries:
    energy_consumption = 0 
    data_s = data[data['country'] == current_country]
    data_s = data_s[data_s['quantity'] > 0]
    data_s = data_s[data_s['year'] >= 2005]
    data_s = data_s[data_s['year'] < 2010]

    data_1 = data_s[data_s['category'] == 4]
    for quan in data_s['quantity']:
        energy_consumption += quan
    data_2 = data_s[data_s['category'] == 12]
    for quan in data_s['quantity']:
        energy_consumption += quan
    data_3 = data_s[data_s['category'] == 21]
    for quan in data_s['quantity']:
        energy_consumption += quan
    print ('Потребление', current_country,'=',energy_consumption) 
```
