Рассчитать количество выбросов для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
import pandas as pd
import numpy as np
#horse = pd.read_csv('horse_data.csv', header=None)
pd.set_option('display.max_columns', 30)
#загрузка данных. Т.к. в файле отсутствует заголовок, передаем names основываясб на файле horse_data.names
horse = pd.read_csv('horse_data.csv', names=['surgery','Age', 'Hospital number', 'rect temp', 'pulse', 'resp rate', 
                                             'temp of extermities', 'periph pulse', 'mucous membranes', 'cap refill time', 
                                             'pain','peristaltisis','abdominal distension','nasogastric tube',
                                             'nasogastric reflux', 'nasogastric reflux PH', 'rectal examination',
                                             'abdomen', 'packed cell volume' ,'total protein','abdominocentesis appearance', 
                                             'abdomcentesis total protein','outcome', 'surgical lesion', '25' ,'26', '27', 
                                             'cp_data'], header=None)

horse.head(20)
```

```python
dtypeCount =[horse.iloc[:,i].apply(type).value_counts() for i in range(horse.shape[1])]
dtypeCount
# проверим типы данных в столбцах до внесения изменений
# сравниваем с описанием в horse_data.names
```

```python
# surgery должен быть integer, пропуск пока заменим на 1
horse['surgery'] = pd.to_numeric(horse['surgery'], errors = 'coerce').fillna(1).astype(np.int64)

# в столбце Age может быть 2 значения (1 и 2), а присутствует 1 и 9. Заменяем на подходящее.
horse.loc[horse['Age'] == 9, 'Age'] = 2

#horse.head(20)
```

```python
#Заменим все ? на NaN
horse = horse.mask(horse == '?')

#а колонку с температурой приведем к соответсвующему её типу float
horse['rect temp'] = horse['rect temp'].astype('float')

#horse.head(30)
#dtypeCount =[horse.iloc[:,i].apply(type).value_counts() for i in range(horse.shape[1])]
#dtypeCount
```

```python
# изменим типы данных в некоторых колонках. Если колонки категориальные, заполним пропуски 0
horse['pulse'] = pd.to_numeric(horse['pulse'], errors = 'coerce').fillna(0).astype(np.int64)
horse['resp rate'] = pd.to_numeric(horse['resp rate'], errors = 'coerce').fillna(0).astype(np.int64)

horse['temp of extermities'] = pd.to_numeric(horse['temp of extermities'], errors = 'coerce').fillna(0).astype(np.int64)
horse['periph pulse'] = pd.to_numeric(horse['periph pulse'], errors = 'coerce').fillna(0).astype(np.int64)

horse['mucous membranes'] = pd.to_numeric(horse['mucous membranes'], errors = 'coerce').fillna(0).astype(np.int64)
horse['cap refill time'] = pd.to_numeric(horse['cap refill time'], errors = 'coerce').fillna(0).astype(np.int64)


horse.head(20)
#horse['surgery'].fillna(1, inplace=True)
#print(horse.iloc[132, :])
```

```python
# посмотрим как изменились типы данных
dtypeCount =[horse.iloc[:,i].apply(type).value_counts() for i in range(horse.shape[1])]
dtypeCount
```

```python
# считаем
horse_filtered = horse[(horse['rect temp'] > 0)]

print('Максимальная ректальная температура =', horse_filtered['rect temp'].max(), 'градусов Цельсия')
print('Минимальная ректальная температура =', horse_filtered['rect temp'].min(), 'градусов Цельсия')
print('Мода ректальной температуры =', horse_filtered['rect temp'].mode()[0], 'градусов Цельсия')
print('Медиана ректальной температуры = ', horse_filtered['rect temp'].median(), 'градусов Цельсия')
horse_filtered.head(10)
```

```python
# тоже самое можно посчитать и для пульса, но тут лучше откинуть минимум и максимум
pulse_low = horse_filtered['pulse'].quantile(0.01)
pulse_hi  = horse_filtered['pulse'].quantile(0.99)

pulse_filtered = horse_filtered[(horse_filtered['pulse'] < pulse_hi) & (horse_filtered['pulse'] > pulse_low)]

print('Максимальый пульс =', pulse_filtered['rect temp'].max(), 'градусов Цельсия')
print('Минимальный пульс =', pulse_filtered['rect temp'].min(), 'градусов Цельсия')
print('Мода пульса =', pulse_filtered['rect temp'].mode()[0], 'градусов Цельсия')
print('Медиана пульса = ', pulse_filtered['rect temp'].median(), 'градусов Цельсия')


#pulse_filtered.head(10)
```

```python
# максимально избавляемся от пропусков данных
horse_final = pulse_filtered[(pulse_filtered['resp rate'] > 0)]
horse_final = horse_final[(horse_final['temp of extermities'] > 0)]
horse_final = horse_final[(horse_final['periph pulse'] > 0)]
horse_final = horse_final[(horse_final['mucous membranes'] > 0)]
horse_final = horse_final[(horse_final['cap refill time'] > 0)]

horse_final.head(20)
```
